# Objective : Perform frequent pattern mining algortihms : Apriori and Fpgrowth

Dataset : This dataset is a groceries data set bought by customers. 

> Our aim is to find the frequent patterns using a minimum threshold using apriori and fpgrowth.

## Uploading the dataset.

```python
from google.colab import files
uploaded = files.upload()
```

> Install the necessary libraries for association rules, apriori and fpgrowth.

```python
import pandas as pd
import io
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import fpgrowth
```

## Dataset looks like below

```python
df=pd.read_csv(io.StringIO(uploaded['groceries - groceries.csv'].decode('utf-8')))
df
```

> This is how the dataset info looks like. It has 33 columns.

```python
df.info()
```

## One hot encoding

> We are dropping the Item(s) column and creating a one hot encoding

```python
basket_sets = pd.get_dummies(df.drop(columns=['Item(s)'],axis=1), prefix="item")
basket_sets = basket_sets.groupby(level=0, axis=1).sum()
basket_sets
```

# Apriori Algorithm

> We are applying the apriori algorithm with min_support as 0.05

> We can see the frequent items with min_support>=0.05 as below. There are 30 itemsets.

```python
frequent_itemsets = apriori(basket_sets, min_support=0.05, use_colnames=True)
frequent_itemsets
```

## Association rules



> Here we are printing the association rules using the lift metric and min_threshold as 1.

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(100)
```

## Patterns

38% of the transactions containing vegetables contain whole milk.

29% of transactions containing whole milk have vegetables.

22% of transactions containing whole milk have rolls/buns.

30% of transactions containing rolls/buns have whole milk.

40% of transactions containing yogurt have whole milk.

21% of transactions containing whole milk have yogurt.

# Fpgrowth

> Now we are appyling the gpgrowth algorithm with min_support as 0.05. We can see the below results with their min_support >=0.05.

```python
freq_items=fpgrowth(basket_sets, min_support=0.05,use_colnames=True)
freq_items
```

## Association rules

> Here we are printing the association rules with confidence as the metric and min_threshold as 0.1. We can see the results below.

```python
rules = association_rules(freq_items, metric="confidence", min_threshold=0.1)
rules.head(5)
```
